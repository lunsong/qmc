```python
import jax
r = jax.random.normal(jax.random.PRNGKey(0), (8,3))
U = lambda r: F24.log_psi(F24.param, r)
F = jax.jit(jax.grad(U))
for step in range(5000):
    r += 1e-3 * F(r)
    if step % 100 == 0:
        print(U(r))
r_24 = r

import jax
r = jax.random.normal(jax.random.PRNGKey(0), (8,3))
U = lambda r: F23.log_psi(F23.param, r)
F = jax.jit(jax.grad(U))
for step in range(5000):
    r += 1e-3 * F(r)
    if step % 100 == 0:
        print(U(r))
r_23 = r
```

```python
import jax
N = 100
xy = jax.numpy.indices((N,N)).reshape(2,-1).T / N - .5

r = (jax.numpy.zeros((N*N,8,3))
     .at[:].set(r_24)
     .at[:,0,jax.numpy.array([0,2])].add(xy))
psi_22 = jax.numpy.exp(jax.vmap(lambda r: F24.log_psi(F24.param, r))(r))

r = (jax.numpy.zeros((N*N,8,3))
     .at[:].set(r_23)
     .at[:,0,jax.numpy.array([0,2])].add(xy))
psi_23 = jax.numpy.exp(jax.vmap(lambda r: F23.log_psi(F23.param, r))(r))


r_24 %= 1.
r_23 %= 1.

plt.close()
idx = jax.numpy.arange(N)/N - .5

ax = plt.subplot(1,2,1,aspect='equal')
# ax.scatter((r_24[:6,0] - r_24[0,0])%1, (r_24[:6,2]-r_24[0,2])%1)
# ax.scatter((r_24[6:,0] - r_24[0,0])%1, (r_24[6:,2]-r_24[0,2])%1)
ax.contour(idx,idx, psi_22.reshape(N,N).T)

ax = plt.subplot(1,2,2,aspect='equal')
# ax.scatter((r_23[:4,0] - r_23[0,0])%1, (r_23[:4,2]-r_23[0,2])%1)
# ax.scatter((r_23[4:,0] - r_23[0,0])%1, (r_23[4:,2]-r_23[0,2])%1)
ax.contour(idx,idx, psi_23.reshape(N,N).T)
```

```python
N = 100

normalize = lambda x: x/jax.numpy.sqrt((x*x).sum())

# n_22 = normalize(r_24[1]-r_24[0])
n_22 = jax.numpy.array([1,0,0])
# n_23 = normalize(r_23[1]-r_23[0])
n_23 = jax.numpy.array([1,0,0])

idx = jax.numpy.arange(N)[:,None]/N

r = (jax.numpy.zeros((N,8,3))
     .at[:].set(r_24)
     .at[:,0].set(idx*r[4]+(1-idx)*r[0]))
     # .at[:,0].set(r_24[4])
     # .at[:,0].add(jax.numpy.arange(N)[:,None]/N * n_22))
psi_22 = jax.numpy.exp(jax.vmap(lambda r: F24.log_psi(F24.param, r))(r))

r = (jax.numpy.zeros((N,8,3))
     .at[:].set(r_23)
     .at[:,0].set(idx*r[4]+(1-idx)*r[0]))
     # .at[:,0].set(r_23[4])
     # .at[:,0].add(jax.numpy.arange(N)[:,None]/N * n_23))
psi_23 = jax.numpy.exp(jax.vmap(lambda r: F23.log_psi(F23.param, r))(r))

psi_23 /= jax.numpy.max(psi_23)
psi_22 /= jax.numpy.max(psi_22)
plt.close()
plt.plot(psi_22,label="22")
plt.plot(psi_23,label="23")
plt.legend()
```
